```python
import numpy as np
from typing import Dict, List, Tuple

def jpeg_compress(im: np.ndarray) -> bytearray:
    """JPEG-подобное сжатие изображения с раздельной обработкой каналов"""
    # 1. Конвертация в YCbCr и даунсэмплинг
    y, cb, cr = func.RGB_to_YCbCr(im)
    cb = func.downsampling(cb)
    cr = func.downsampling(cr)

    # 2. Разбиение на блоки
    y_blocks = func.split_into_blocks(y)
    cb_blocks = func.split_into_blocks(cb)
    cr_blocks = func.split_into_blocks(cr)

    bitstream = ''  # весь поток накапливается здесь
    prev_dc = {'y': 0, 'Cb': 0, 'Cr': 0}
    
    ha_tables = {
        'y': {'dc': HA_Y_DC_TABLE, 'ac': HA_Y_AC_TABLE},
        'Cb': {'dc': HA_C_DC_TABLE, 'ac': HA_C_AC_TABLE},
        'Cr': {'dc': HA_C_DC_TABLE, 'ac': HA_C_AC_TABLE}
    }

    # 3. Обработка блоков
    for channel, blocks, q_matrix in [('y', y_blocks, Q_Y),
                                      ('Cb', cb_blocks, Q_C),
                                      ('Cr', cr_blocks, Q_C)]:
        for row in blocks:
            for block in row:
                # 4. DCT и квантование
                dct_block = func.dct_2d(block - 128)
                quant_block = func.quantize(dct_block, q_matrix)
                zz_coeffs = func.zigzag(quant_block)

                # 5. Кодирование DC
                dc_value = zz_coeffs[0]
                dc_diff = dc_value - prev_dc[channel]
                prev_dc[channel] = dc_value

                dc_size = 0 if dc_diff == 0 else int(np.floor(np.log2(abs(dc_diff))) + 1)
                dc_code = ha_tables[channel]['dc'].get(dc_size, '')
                if dc_code == '':
                    raise ValueError(f"No DC Huffman code for size {dc_size}")

                dc_bits = dc_code
                if dc_size > 0:
                    if dc_diff >= 0:
                        dc_bits += bin(dc_diff)[2:].zfill(dc_size)
                    else:
                        dc_bits += bin((1 << dc_size) + dc_diff - 1)[2:]

                # 6. Кодирование AC
                ac_coeffs = zz_coeffs[1:]
                ac_rle = func.rle_ac(ac_coeffs)
                ac_bits = ''
                for run, value in ac_rle:
                    if (run, value) == (0, 0):
                        ac_bits += ha_tables[channel]['ac'].get((0, 0), '')
                        break

                    size = int(np.floor(np.log2(abs(value))) + 1) if value != 0 else 0
                    ac_code = ha_tables[channel]['ac'].get((run, size), '')
                    if ac_code == '':
                        raise ValueError(f"No AC Huffman code for (run={run}, size={size})")
                    
                    ac_bits += ac_code
                    if size > 0:
                        if value > 0:
                            ac_bits += bin(value)[2:].zfill(size)
                        else:
                            ac_bits += bin((1 << size) + value - 1)[2:]

                # В случае отсутствия EOB — явно добавим
                if (0, 0) not in ac_rle:
                    ac_bits += ha_tables[channel]['ac'].get((0, 0), '')

                # 7. Добавим в общий битовый поток
                bitstream += dc_bits + ac_bits

    # 8. Паддинг до байтов
    pad_len = (8 - len(bitstream) % 8) % 8
    bitstream += '0' * pad_len

    # 9. Преобразование в байты
    comp_data = bytearray(int(bitstream[i:i+8], 2) for i in range(0, len(bitstream), 8))
    return comp_data
```

```python
import numpy as np
from typing import Tuple, Dict

def jpeg_decompress(compressed_data: bytearray, original_shape: Tuple[int, int, int],
                   Q_Y: np.ndarray, Q_C: np.ndarray) -> np.ndarray:
    """Устойчивый декомпрессор с обработкой ошибок"""
    # Инициализация каналов
    height, width, _ = original_shape
    y_channel = np.full((height, width), 128, dtype=np.float32)
    cb_channel = np.full((height//2, width//2), 128, dtype=np.float32)
    cr_channel = np.full((height//2, width//2), 128, dtype=np.float32)
    
    # Конвертация в битовую строку
    bitstream = ''.join(f'{byte:08b}' for byte in compressed_data)
    
    # Обработка каждого канала
    for channel_name, channel, q_matrix in [
        ('y', y_channel, Q_Y),
        ('Cb', cb_channel, Q_C),
        ('Cr', cr_channel, Q_C)
    ]:
        prev_dc = 0
        blocks_h = (channel.shape[0] + 7) // 8
        blocks_w = (channel.shape[1] + 7) // 8
        bit_pos = 0
        
        for block_idx in range(blocks_h * blocks_w):
            if bit_pos + 16 > len(bitstream):  # Минимальный размер для DC
                break
                
            # Декодирование DC
            dc_diff, bits_used = decode_huffman(bitstream[bit_pos:], DC_INV_TABLES[channel_name], is_dc=True)
            bit_pos += bits_used
            dc_value = prev_dc + dc_diff
            prev_dc = dc_value
            
            # Декодирование AC
            ac_coeffs, bits_used = decode_ac(bitstream[bit_pos:], AC_INV_TABLES[channel_name])
            bit_pos += bits_used
            
            # Восстановление блока
            zz_coeffs = [dc_value] + ac_coeffs
            print(len(zz_coeffs))
            quant_block = inverse_zigzag(zz_coeffs)
            dct_block = quant_block * q_matrix
            block = idct_2d(dct_block) + 128
            
            # Размещение блока
            i, j = divmod(block_idx, blocks_w)
            y, x = i * 8, j * 8
            h, w = channel.shape
            block_h, block_w = min(8, h-y), min(8, w-x)
            channel[y:y+block_h, x:x+block_w] = block[:block_h, :block_w]
    
    # Восстановление цветового пространства
    cb_upsampled = np.repeat(np.repeat(cb_channel, 2, axis=0), 2, axis=1)[:height, :width]
    cr_upsampled = np.repeat(np.repeat(cr_channel, 2, axis=0), 2, axis=1)[:height, :width]
    
    rgb = ycbcr_to_rgb(y_channel, cb_upsampled, cr_upsampled)
    return np.clip(rgb, 0, 255).astype(np.uint8)

DC_INV_TABLES = {
    'y': {
        '00': 0,        # Категория 0 (0)
        '010': 1,       # Категория 1 (+1)
        '011': 2,       # Категория 2 (+2)
        '100': 3,       # Категория 3 (+3)
        '101': 4,       # Категория 4 (+4)
        '110': 5,       # Категория 5 (+5)
        '1110': 6,      # Категория 6 (+6)
        '11110': 7,     # Категория 7 (+7)
        '111110': 8,    # Категория 8 (+8)
        '1111110': 9,   # Категория 9 (+9)
        '11111110': 10  # Категория 10 (+10)
    },
    'Cb': {
        '00': 0,        # Категория 0 (0)
        '010': 1,       # Категория 1 (+1)
        '011': 2,       # Категория 2 (+2)
        '100': 3,       # Категория 3 (+3)
        '101': 4,       # Категория 4 (+4)
        '110': 5,       # Категория 5 (+5)
        '1110': 6,      # Категория 6 (+6)
        '11110': 7,     # Категория 7 (+7)
        '111110': 8,    # Категория 8 (+8)
        '1111110': 9,   # Категория 9 (+9)
        '11111110': 10  # Категория 10 (+10)
    },
    'Cr': {
        '00': 0,        # Категория 0 (0)
        '010': 1,       # Категория 1 (+1)
        '011': 2,       # Категория 2 (+2)
        '100': 3,       # Категория 3 (+3)
        '101': 4,       # Категория 4 (+4)
        '110': 5,       # Категория 5 (+5)
        '1110': 6,      # Категория 6 (+6)
        '11110': 7,     # Категория 7 (+7)
        '111110': 8,    # Категория 8 (+8)
        '1111110': 9,   # Категория 9 (+9)
        '11111110': 10  # Категория 10 (+10)
    }
}


AC_INV_TABLES = {
    'y': {
        '1010': (0, 0),   # (run=0, size=0)
        '00': (0, 1),     # (run=0, size=1)
        '01': (0, 2),     # (run=0, size=2)
        '10': (0, 3),     # (run=0, size=3)
        '110': (0, 4),    # (run=0, size=4)
        '1110': (0, 5),   # (run=0, size=5)
        '11110': (0, 6),  # (run=0, size=6)
        '111110': (0, 7), # (run=0, size=7)
        '1111110': (0, 8),# (run=0, size=8)
        '11111110': (0, 9),# (run=0, size=9)
        '111111110': (0, 10)  # (run=0, size=10)
    },
    'Cb': {
        '1010': (0, 0),   # (run=0, size=0)
        '00': (0, 1),     # (run=0, size=1)
        '01': (0, 2),     # (run=0, size=2)
        '10': (0, 3),     # (run=0, size=3)
        '110': (0, 4),    # (run=0, size=4)
        '1110': (0, 5),   # (run=0, size=5)
        '11110': (0, 6),  # (run=0, size=6)
        '111110': (0, 7), # (run=0, size=7)
        '1111110': (0, 8),# (run=0, size=8)
        '11111110': (0, 9),# (run=0, size=9)
        '111111110': (0, 10)  # (run=0, size=10)
    },
    'Cr': {
        '1010': (0, 0),   # (run=0, size=0)
        '00': (0, 1),     # (run=0, size=1)
        '01': (0, 2),     # (run=0, size=2)
        '10': (0, 3),     # (run=0, size=3)
        '110': (0, 4),    # (run=0, size=4)
        '1110': (0, 5),   # (run=0, size=5)
        '11110': (0, 6),  # (run=0, size=6)
        '111110': (0, 7), # (run=0, size=7)
        '1111110': (0, 8),# (run=0, size=8)
        '11111110': (0, 9),# (run=0, size=9)
        '111111110': (0, 10)  # (run=0, size=10)
    }
}


def decode_huffman(bitstream: str, inv_table: Dict, is_dc: bool) -> Tuple[int, int]:
    """Декодирование одного значения Хаффмана"""
    code = ''
    for bit in bitstream:
        code += bit
        if code in inv_table:
            # Обработка DC
            if is_dc:
                category = inv_table[code]
                if category == 0:
                    return 0, len(code)  # Разность DC = 0
                # Чтение дополнительных битов для значения DC
                bits_needed = category
                remaining_bits = bitstream[len(code):len(code) + bits_needed]
                if len(remaining_bits) < bits_needed:
                    raise ValueError("Недостаточно битов для DC коэффициента")
                value = int(remaining_bits, 2)  # Положительное или отрицательное значение
                if remaining_bits[0] == '1':  # Если отрицательное
                    value = value - (1 << bits_needed) + 1
                return value, len(code) + bits_needed
            else:
                # Обработка AC
                run, size = inv_table[code]
                return run, size, len(code)
    
    #print(f"Не найден код Хаффмана для кода: {code}, оставшийся битовый поток: {bitstream}")
    raise ValueError("Не найден код Хаффмана")

def decode_ac(bitstream: str, inv_ac_table: Dict) -> Tuple[list, int]:
    """Декодирование AC коэффициентов"""
    ac_coeffs = []
    total_bits = 0
    zeros_to_add = 0
    
    while len(ac_coeffs) + zeros_to_add < 63:
        # Декодируем пару (run, size)
        run, size, bits_used = decode_huffman(bitstream, inv_ac_table, is_dc=False)
        bitstream = bitstream[bits_used:]
        total_bits += bits_used
        
        if run == 0 and size == 0:  # EOB (End Of Block)
            ac_coeffs.extend([0] * (63 - len(ac_coeffs)))  # Заполнение нулями до 63
            break
        
        if run == 15 and size == 0:  # ZRL (Zero Run Length)
            zeros_to_add += 16
            continue
        
        # Добавляем нули в соответствии с run
        ac_coeffs.extend([0] * run)
        
        if size > 0:
            bits = bitstream[:size]
            bitstream = bitstream[size:]
            total_bits += size
            
            if bits[0] == '1':  # Положительное
                value = int(bits, 2)
            else:  # Отрицательное
                value = int(bits, 2) - (1 << size) + 1
            
            ac_coeffs.append(value)
        else:
            ac_coeffs.append(0)
    
    return ac_coeffs[:63], total_bits

def inverse_zigzag(zz_coeffs) -> np.ndarray:
    """Обратное зигзаг-преобразование"""
    block = np.zeros((8, 8))
    zigzag_order = [
         0,  1,  5,  6, 14, 15, 27, 28,
         2,  4,  7, 13, 16, 26, 29, 42,
         3,  8, 12, 17, 25, 30, 41, 43,
         9, 11, 18, 24, 31, 40, 44, 53,
        10, 19, 23, 32, 39, 45, 52, 54,
        20, 22, 33, 38, 46, 51, 55, 60,
        21, 34, 37, 47, 50, 56, 59, 61,
        35, 36, 48, 49, 57, 58, 62, 63
    ]
    for i, pos in enumerate(zigzag_order):
        if i < len(zz_coeffs):
            block[pos//8, pos%8] = zz_coeffs[i]
    return block

def idct_2d(block: np.ndarray) -> np.ndarray:
    """Обратное DCT преобразование"""
    idct_block = np.zeros((8, 8))
    for x in range(8):
        for y in range(8):
            sum_val = 0
            for u in range(8):
                for v in range(8):
                    cu = 1/np.sqrt(2) if u == 0 else 1
                    cv = 1/np.sqrt(2) if v == 0 else 1
                    sum_val += (
                        cu * cv * block[u,v] * 
                        np.cos((2*x+1)*u*np.pi/16) * 
                        np.cos((2*y+1)*v*np.pi/16)
                    )
            idct_block[x,y] = 0.25 * sum_val
    return idct_block

def ycbcr_to_rgb(y: np.ndarray, cb: np.ndarray, cr: np.ndarray) -> np.ndarray:
    """Конвертация YCbCr в RGB"""
    r = y + 1.402 * (cr - 128)
    g = y - 0.34414 * (cb - 128) - 0.71414 * (cr - 128)
    b = y + 1.772 * (cb - 128)
    return np.stack([r, g, b], axis=-1)
```
